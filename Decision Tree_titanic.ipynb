```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
df = pd.read_csv('titanic_main.csv')
df.head()
```

```python
ndf = df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis=1)
le = LabelEncoder()
ndf["sex"]=le.fit_transform(ndf["Sex"])
ndf = ndf.drop("Sex",axis=1)
x = ndf.drop("Survived",axis=1)
x = x.fillna(0)
y = ndf["Survived"]
features = ["Pclass","sex","Age","Fare"]
model = DecisionTreeClassifier()
model.fit(x,y)
model.score(x,y)
model.predict([[1,0,10,71.25]])
tree.plot_tree(model,feature_names = features)
```
