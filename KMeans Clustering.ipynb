```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
odf = pd.read_csv("income.csv") 
df = odf.drop("Name",axis=1)
df.head()

# Old Clusters
plt.scatter(df.Age,df.Income)
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()

# Old Prediction
km = KMeans(n_clusters=3)
pre = km.fit_predict(df[["Age","Income"]])
df['cluster'] = pre
df.head()

plt.scatter(df.Age,df.Income)
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()

df0 = df[df.cluster==0]
df1 = df[df.cluster==1]
df2 = df[df.cluster==2]

plt.scatter(df0.Age,df0.Income,label = "Income", color ='red')
plt.scatter(df1.Age,df1.Income,label = "Income",  color ='green')
plt.scatter(df2.Age,df2.Income,label = "Income",  color ='black')
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
```

```python
#  Preprossing the data
# Scaling the data
scaler = MinMaxScaler()
scaler.fit(df[['Income']])
df['Income'] = scaler.transform(df[['Income']])
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df.head()
```

```python
plt.scatter(df.Age,df.Income)
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()
```

```python
km = KMeans(n_clusters=3)
pre = km.fit_predict(df[["Age","Income"]])
df['cluster'] = pre
df.head()
```

```python
plt.scatter(df.Age,df.Income)
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()

df0 = df[df.cluster==0]
df1 = df[df.cluster==1]
df2 = df[df.cluster==2]

plt.scatter(df0.Age,df0.Income,label = "Income", color ='red')
plt.scatter(df1.Age,df1.Income,label = "Income",  color ='green')
plt.scatter(df2.Age,df2.Income,label = "Income",  color ='black')
plt.xlabel("Age")
plt.ylabel("Income")
plt.legend()

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker='*',c='purple',label='centroid')
```

```python
# using Elbow Method
sse =[]
krng = range(1,11)
for i in krng:
    km = KMeans(n_clusters=i)
    km.fit(df[["Age","Income"]])
    sse.append(km.inertia_)
plt.plot(krng,sse)
plt.xlabel("K")
plt.ylabel("Sum of Squered Error")
plt.legend()
plt.show
```
