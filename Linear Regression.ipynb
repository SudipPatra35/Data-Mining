```python
import numpy as np
import pandas as pd
import matplotlib as mt
from matplotlib import pyplot as plt
from sklearn import linear_model
df = pd.read_csv('homeprices.csv')
```

```python
# scatter ploting
plt.scatter(df.area,df.price)
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
```

```python
x = df.drop('price',axis=1)
y =df['price']
model = linear_model.LinearRegression()
model.fit(x,y)
# pre = model.predict([y])
model.predict([[3500]])
```

```python
n_df = pd.read_csv('areas.csv')
pre = model.predict(n_df)
n_df['price'] = pre
n_df.to_csv('AreasPrices.csv')
```
